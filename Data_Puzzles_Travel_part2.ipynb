# Part 2: Can you get the two nearest distinct capitals on Earth? (sea level)

```python
import numpy as np
import pandas as pd
from geopy import distance
```

```python
capitals_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/features/travel/worldcapitals_light.csv')
```

```python
capitals_df.head()
```

```python
# again, we use geopy distance library, we need to consider distinct capitals.
```

```python
def calc_dist(city1, city2):
  city1_coords = (city1.loc['lat'], city1.loc['lng'])
  city2_coords = (city2.loc['lat'], city2.loc['lng'])
  return distance.distance(city1_coords, city2_coords).kilometers
```

```python
min_dist = float("inf")
best_pair = ()
for i in range(len(capitals_df)):
  for j in range(i + 1, len(capitals_df)): # i + 1 means we keep it distinct
    city1 = capitals_df.iloc[i]
    city2 = capitals_df.iloc[j]
    test_dist = calc_dist(city1, city2)
    if (test_dist < min_dist):
      min_dist = test_dist
      best_pair = (city1.loc['city'], city2.loc['city'])
(best_pair, min_dist)
```

```python
# Al-Quds is another name for Jerusalem, so we should remove this from consideration.

# in real world datasets may contain tricks like this, so we have to watch out
```
