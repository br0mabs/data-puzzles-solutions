```python
# the approach is to visualize the speeds limits for each road, and then we can check the clustering to see the speed limit

# filter dataframe by each road name, generate a plot for each in a for loop, then read them off, since there are only 6
```

```python
for road in speeds_df['road'].unique():
  road_df = speeds_df[speeds_df['road'] == road]
  # now plot speeds
  # alternate idea without visualizing: use mean
  # a good visualization can be histogram with 1 unit buckets
  plt.hist(road_df['speed'], bins=100)
  plt.title(road)
  plt.show()
```

```python
# it is common knowledge that speed limits are multiples of 10, so based on that information:
ans = 50 + 70 + 100 + 90 + 130 + 30
ans
```

```python
# note that the averages (rounded to the nearest 10) are instead:
for road in speeds_df['road'].unique():
  road_df = speeds_df[speeds_df['road'] == road]
  road_mean = round(road_df['speed'].mean(), -1)
  print(road, road_mean)

# which is not what we expect from the visualization (for R77)
# so this is why visualizing the data is usually always a better idea
```
